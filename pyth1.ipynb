```
import time
```

```
import math
```

```
print(math.sqrt(100))
```

```
vals=list(range(1,100))
tic=time.time()
for x in vals:
  y=1
  for i in range(1,x):
    y *= i
toc=time.time()
print('Elapsed own ', toc - tic)  
tic=time.time()
for x in vals:
  y=math.factorial(x)
toc=time.time()
print('Elapsed math ', toc - tic) 
```
